# CBSC Strategy Workspace - Introduction

This notebook demonstrates the basic usage of the CBSC Strategy Workspace for strategy development in Jupyter notebooks.

## Overview

The `StrategyWorkspace` class provides a unified interface for:
- Fetching historical market data (OHLCV)
- Managing available trading symbols
- Caching data for improved performance
- Session state management

## 1. Import and Initialize Workspace

The workspace should be used as an async context manager to ensure proper resource handling.

```python
from cbsc_strategy_sdk import StrategyWorkspace
from datetime import date
import nest_asyncio

# Enable nested async for Jupyter
nest_asyncio.apply()

# Initialize workspace with default settings
ws = StrategyWorkspace(
    api_base="http://localhost:3003",
    cache_ttl=300,  # 5 minutes
    timeout=30
)
```

## 2. Use Async Context Manager

Enter the async context to initialize the workspace and its data connector.

```python
import asyncio

async def init_workspace():
    async with ws:
        print(f"Workspace initialized: {ws}")
        print(f"Config: {ws.config}")
        return ws

# Run initialization
ws = asyncio.run(init_workspace())
```

## 3. Fetch Historical Data

Retrieve OHLCV (Open, High, Low, Close, Volume) data for a specific symbol and date range.

```python
# Define date range
start_date = date(2024, 1, 1)
end_date = date(2024, 1, 31)

# Fetch historical data for Apple Inc.
async def fetch_data():
    async with ws:
        data = ws.get_historical_data(
            symbol="AAPL",
            start=start_date,
            end=end_date,
            interval="1d"
        )
        return data

aapl_data = asyncio.run(fetch_data())

# Display first few rows
print("Sample AAPL Data:")
print(aapl_data.head())
```

## 4. Explore the Data

The returned data is a pandas DataFrame with columns: `open`, `high`, `low`, `close`, `volume`

```python
# Display data info
print("
DataFrame Info:")
print(aapl_data.info())

# Display statistics
print("
Descriptive Statistics:")
print(aapl_data.describe())
```

## 5. Get Available Symbols

Retrieve the list of all available trading symbols from the CBSC backend.

```python
async def get_symbols():
    async with ws:
        symbols = ws.get_available_symbols()
        return symbols

symbols = asyncio.run(get_symbols())

print(f"Total available symbols: {len(symbols)}")
print("
First 10 symbols:")
for symbol in symbols[:10]:
    print(f"  - {symbol}")
```

## 6. Cache Management

The workspace automatically caches data to improve performance. Check cache status and clear if needed.

```python
async def check_cache():
    async with ws:
        print(f"Current cache size: {ws.cache_size} entries")
        print(f"Cache TTL: {ws.config.cache_ttl} seconds")
        print(f"Cache enabled: {ws.config.is_cache_enabled()}")

asyncio.run(check_cache())
```
