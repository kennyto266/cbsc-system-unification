# CBSC Strategy SDK - Visualization Module Demo

This notebook demonstrates the visualization capabilities of the CBSC Strategy SDK.

## 1. Setup and Imports

```python
import sys
sys.path.insert(0, '..')

from cbsc_strategy_sdk.visualization import (
    ChartBuilder,
    PerformanceCharts,
    get_theme,
    list_themes,
    create_sample_data,
)

print('Available themes:', list_themes())
```

## 2. Create Sample Data

```python
# Generate sample OHLCV data
data = create_sample_data(days=100)
print(f'Generated {len(data)} days of OHLCV data')
data.head()
```

## 3. Candlestick Chart with Indicators

```python
# Create candlestick chart
builder = ChartBuilder(theme='dark')
fig = builder.create_candlestick(data, title='Price Chart')

# Add moving averages
fig = builder.add_moving_average(data, periods=[5, 10, 20])

# Add Bollinger Bands
fig = builder.add_bollinger_bands(data)

fig.show()
```

## 4. Performance Analysis

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Generate sample returns
dates = pd.date_range(start=datetime.now() - timedelta(days=252), periods=252, freq='D')
returns = pd.Series(np.random.normal(0.001, 0.02, 252), index=dates)

# Performance charts
perf = PerformanceCharts(theme='light')
fig = perf.plot_performance(returns)
fig.show()
```

## 5. Export Chart

```python
# Export to HTML
builder.export_html('my_chart.html')
print('Chart exported!')
```
