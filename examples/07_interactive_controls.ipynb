# Interactive Controls for Strategy Management

This notebook demonstrates the interactive controls module.

```python
from cbsc_strategy_sdk import (
    ControlWidgets,
    SymbolSelector,
    DateRangePicker,
    ParameterValidator,
    QuickControlPanel
)
from datetime import datetime
import ipywidgets as widgets
from IPython.display import display

print('Imports successful')
```

## 1. Basic Widgets

```python
slider = ControlWidgets.slider('test', 0, 100, 50, description='Test:')
dropdown = ControlWidgets.dropdown('choice', ['A', 'B', 'C'], 'A')
checkbox = ControlWidgets.checkbox('enabled', True)

display(widgets.VBox([slider, dropdown, checkbox]))
```

## 2. Symbol Selector

```python
selector = SymbolSelector(['AAPL', 'MSFT', 'GOOGL', 'AMZN'], max_selections=3)
selector.display()
```

## 3. Date Range Picker

```python
picker = DateRangePicker()
picker.display()
```

## 4. Parameter Validation

```python
validator = ParameterValidator()
validator.add_range_rule('value', min_value=0, max_value=100)

result = validator.validate({'value': 50})
display(result.to_widget())
print(f'Valid: {result.is_valid}')
```

## 5. Quick Control Panel

```python
params = {
    'param1': {'type': 'slider', 'min': 0, 'max': 100, 'initial': 50},
    'param2': {'type': 'dropdown', 'choices': ['A', 'B'], 'initial': 'A'}
}

panel = QuickControlPanel(params)
panel.display()
print('Values:', panel.get_values())
```
