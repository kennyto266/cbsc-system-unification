# Simple Moving Average Strategy


```python
import pandas as pd
import numpy as np

# Configuration
SYMBOL = 'AAPL'
SHORT_MA = 5
LONG_MA = 10

print(f'Testing {SYMBOL} with MA({SHORT_MA}) and MA({LONG_MA)})')
```

```python
# Generate sample price data
np.random.seed(42)
prices = np.random.randn(20).cumsum() + 100
dates = pd.date_range('2024-01-01', periods=20)

data = pd.DataFrame({
    'date': dates,
    'close': prices
}).set_index('date')

print(data.head())
```

```python
# Calculate moving averages
data['ma_short'] = data['close'].rolling(SHORT_MA).mean()
data['ma_long'] = data['close'].rolling(LONG_MA).mean()

# Generate signals
data['signal'] = 0
data.loc[data['ma_short'] > data['ma_long'], 'signal'] = 1

print('Signals generated:')
print(data[['close', 'ma_short', 'ma_long', 'signal']].tail())
```

```python
# Calculate returns
data['returns'] = data['close'].pct_change()

# Strategy return (next day return if signal is 1)
data['strategy_return'] = data['returns'].shift(-1) * data['signal']

total_return = data['strategy_return'].sum()
print(f'Total Strategy Return: {total_return:.2%}')
```
